?1) Create Sex_num column using replace in train

```python
train['Sex_num'] = train.Sex.replace({'male':0,'female':1})
```

```python
train.loc[:,['Sex','Sex_num']]
```

?2) Find the length of all elements and put those in a dataframe for corey_df

```python
corey_df.loc[0:2,['first','last','email']].applymap(len)
```

?3) Create a column 'full_name' by combining the first and last name columns and drop the first and last colums for corey_df

```python
corey_df['full_name'] = corey_df['first'] + ' ' + corey_df['last']
```

```python
corey_df.head()
```

```python
corey_df.drop(['first','last'],axis=1,inplace=True)
```

?4) Create columns first and last columns by splitting full_name column for corey_df using expand.

```python
corey_df[['first','last']] = corey_df['full_name'].str.split(' ',expand=True)
```

```python
corey_df.head()
```

?5) Add Tony to column first and nothing else.

```python
corey_df.append({'first':'Tony'},ignore_index=True)
```

?6) Create a new dataframe for Tony Stark with email id tony@avenge.com and add that to corey_df

```python
tony_df = pd.DataFrame({'first':'Tony','last':'Stark','email':'tony@avenge.com'},index=[0])
```

```python
tony_df
```

```python
corey_df = corey_df.append(tony_df,ignore_index=True,sort=False)
```

?7) Drop the Columns with last name Doe in corey_df

```python
corey_df.drop(index=corey_df[corey_df['last']=='Doe'].index)
```

?12) Convert the date column into actual date and time format

https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

```python
eth['Date'] = pd.to_datetime(eth.Date,format='%Y-%m-%d %I-%p')
```

```python
eth.dtypes
```

?13) Print the name of the day of the first observation

```python
eth.loc[0,'Date'].day_name()
```

?14) Display the data only for 2017

```python
eth.loc[eth.Date.dt.year==2017,:]
```
