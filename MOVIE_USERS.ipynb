```python
users = pd.read_table('http://bit.ly/movieusers',sep='|',header=None,names=['user id','age','gender','occupation','zip code'])
```

```python
users.head()
```

18.  Define a function to_upper for converting the values in occupation to capital values. Execute the same


```python
def to_upper(col):
  return col.upper()
```

```python
users['OCCUPATION'] = users.occupation.apply(to_upper)
```

```python
users.head()
```

19.  Do the operation in 18 using lambda function

```python
OC = lambda col : col.upper()
```

```python
users['OCCU'] = users.occupation.apply(OC)
```
