8. Create two dataframes, test and study from the train dataframe. test should have 66% of randomly selected rows of train and rest should be in study.No overlapping permitted.

```python
test = train.sample(frac=0.66)
```

```python
study = train.loc[~train.index.isin(test.index),:]
```

**RESET THE DATAFRAME**
    
    

9. Create a new column Sex_num using map method. Females should have 1 and males should have 0 in Sex_num.
    

```python
train['Sex_num'] = train.Sex.map({'female':1,'male':0})
```

```python
train.head()
```

10. Create a new column Name_length with the length of the entries in Name column using apply
    

```python
train['Name_length'] = train.Name.apply(len)
```

```python
train.head()
```

11. Create a column with the last Names of people using apply.
   

```python
train['last'] = train.Name.apply(lambda nam:nam.split(',')[0])
```

```python
train.head()
```

 12. Take the columns PassengerId,Survived and Pclass and display the largest entry row-wise using apply
    

```python
train.loc[:,['PassengerId','Survived','Pclass']].apply(max,axis=1)
```

13.  Take the columns PassengerId,Survived and Pclass and display the largest entry column-wise using apply
    

```python
train.loc[:,['PassengerId','Survived','Pclass']].apply(max,axis=0)
```

14. Take the columns PassengerId,Survived and Pclass and display the index of column with the largest entry row-wise using apply
    

```python
import numpy as np
```

```python
train.loc[:,['PassengerId','Survived','Pclass']].apply(np.argmax,axis=1)
```

15. Take the columns PassengerId,Survived and Pclass and change all the numerical entries to float

```python
train.loc[:,['PassengerId','Survived','Pclass']].applymap(float)
```
