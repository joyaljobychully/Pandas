?2) In train1, rename the PassengerId column as PId and merge the two dataframes

```python
train1.rename({'PassengerId':'PId'},axis=1,inplace=True)
```

```python
train1.head()
```

```python
pd.merge(train1,train2,left_on='PId',right_on='PassengerId')
```

?3) Set the column PId as the index column of train1 and merge train1 and train2

```python
train1.set_index('PId',inplace=True)
```

```python
pd.merge(train1,train2,left_index=True,right_on='PassengerId')
```

##**Index from the left dataframe became a column in the merged dataframe and the index of the merged dataframe came from the dataframe on the right**

?4) Change the index of train2 to PassengerId and merge train1 and train2

```python
train2.set_index('PassengerId',inplace=True)
```

```python
pd.merge(train1,train2,left_index=True,right_index=True)
```

##**The left index was used now as the index of the merged dataframe**
